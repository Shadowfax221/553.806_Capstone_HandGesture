## Data Preprocessing

#### GrayScaling and Resizing

```python
from PIL import Image
import os
import shutil
import matplotlib.pyplot as plt
import numpy as np
```

```python
source_dir = 'dataset'


target_dir = 'processed_dataset'
os.makedirs(target_dir, exist_ok=True)


new_size = (512, 512) 


for subdir in os.listdir(source_dir):
    source_subdir = os.path.join(source_dir, subdir)

    if os.path.isdir(source_subdir):
        target_subdir = os.path.join(target_dir, subdir)
        os.makedirs(target_subdir, exist_ok=True)

        for file_name in os.listdir(source_subdir):
            if file_name.endswith('.jpg'):

                img_path = os.path.join(source_subdir, file_name)
                img = Image.open(img_path)

                resized_img = img.resize(new_size)
                gray_img = resized_img.convert('L')

                gray_img_path = os.path.join(target_subdir, file_name)
                gray_img.save(gray_img_path)
```

```python
processed_dir = 'processed_dataset'


for subdir in os.listdir(processed_dir):
    subdir_path = os.path.join(processed_dir, subdir)
    if os.path.isdir(subdir_path):

        first_image_name = next((f for f in os.listdir(subdir_path) if f.endswith('.jpg')), None)
        if first_image_name:
            img_path = os.path.join(subdir_path, first_image_name)
            img = Image.open(img_path)


            plt.imshow(img, cmap='gray')
            plt.title(subdir)
            plt.show()


            img_array = np.array(img)
            print(f"Image array for {subdir}: \n{img_array}\n")
            print(f"Shape of the image array: {img_array.shape}\n")
```
